```python


from src.factories.datasets import dataloader as dl
from src.factories.datasets import animal_10 as ds
from src.factories.models import vanilla_vae as vae
```

```python
dataloader = dl.animal_data_loader()
model = vae.VanillaVAE()
```

```python

img_tensor, label = dataloader.dataset[1000]
print(f"Image shape: {img_tensor.shape}")
print(f"Label: {label} ({dataloader.dataset.class_names[label]})")

# Visualize
ds.show_tensor(img_tensor)
```

```python
import random
import torch

def sample(output_tensor:torch.Tensor):
    random_index = random.randint(0, output_tensor.shape[0])
    return output_tensor[random_index,:,:]
```

```python
for batch in dataloader:
    inputs, _ = batch
    output = model(inputs)
    sample_output = sample(output)
    ds.show_tensor(sample_output)
    break
```
